# Tree Figure on Iris

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import datasets
from numpy.random import permutation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define theme
sns.set_theme(
    color_codes=True, palette="bright", style="white", context="talk", font_scale=1.5
)

# Load dataset
iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target

# Randomize samples
p = permutation(iris_X.shape[0])
iris_X = iris_X[p]
iris_y = iris_y[p]

fig, ax = plt.subplots(
    nrows=1, ncols=2, figsize=(30, 15), gridspec_kw={"width_ratios": [1, 2]}
)

# Update the tree and plot tree structures
sdt = DecisionTreeClassifier()
sdt.partial_fit(iris_X[:25], iris_y[:25], classes=[0, 1, 2])
sdt_plot = plot_tree(
    sdt,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    ax=ax[0],
)
sdt.partial_fit(iris_X[25:50], iris_y[25:50])
sdt_plot = plot_tree(
    sdt,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    ax=ax[1],
)
plt.savefig("../paper/iris.pdf", transparent=True, bbox_inches="tight")
```
